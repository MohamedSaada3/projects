```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('tmdb-movies.csv')
df
#im exploring the shape of the data


```

```python
df['profit']=df['revenue']-df['budget']
#here i wanted to have a specific column that has the profit of each movie 
# it is a fact the profit gives more accurate information about the movie's success
#this will help me in later visualization to determine if a certain movie genre has profit growth
df.head()
```

```python
from matplotlib.pyplot import figure

x=df["popularity"].head(5)
y=df['original_title'].head(5)
figure(figsize=(12, 10), dpi=80)

ax = plt.bar(y,x, color ='maroon', 
        width = 0.5,)



plt.xlabel("movie title")
plt.ylabel("popularity")
plt.title("top 5 movies in popularity")
plt.show()
```



**my first question that i want to ask is: what is the top  movies in terms of popularity**
this may help me in discovering more info about the data 
i choose to order it by popularity because later i want to see if there is a correlation between popularity and profit

```python
df=df.sort_values('popularity',ascending=False) 
# i sorted the values in terms of popularity 
#this step mainly helped me to discover more information about the data
df.reset_index(inplace=True) 
#here i used reset index to make the index right since i reordered the data 
df




```

**analyzing the data of the sci-fi movies**


```python
import pandas as pd

def filter_dataframe_by_string(string_to_find):
    
    df1 = df[['original_title','genres','release_year','profit','popularity']]


    # Handling missing values: Replace NaN or None with an empty string
    df1['genres'] = df['genres'].fillna('')

    # Filter rows containing the specified string in the 'genres' column
    filtered_df = df1[df1['genres'].str.contains(string_to_find)]
    
    return filtered_df

#plz make sure to write the name of the genre correct (its case sensitive)
string_to_find = 'Comedy'
df_filtered = filter_dataframe_by_string(string_to_find)
df_filtered.sort_values('profit', ascending=False).head(5)
print(df_filtered['profit'].mean())














```

**now we will visualize the profit of the choosen genre movies over the years**

```python
plt.bar(df_filtered['release_year'], df_filtered['profit'], color ='maroon', width = 0.7)
plt.title(f'the growth of the {string_to_find} movies over the past years')
plt.ylabel("profit")
plt.xlabel("release year")
plt.show()

```

**now lets see if popularity and profit effects eachother (has a correlation)**

```python
x=df['popularity'].corr(df['profit'])
print(x)
```

**here is the visualization of the correlation between popularity and profit**

```python
sns.heatmap(df[['popularity', 'profit']].corr())
```

**another question that i want to answer is whether the choosen column effects the profit (negatively or positively)**

```python
def profit_scatter(column):
    x=df['profit']
    y=df[column]
    plt.scatter(x,y,c='green')
    plt.ylabel(column)
    plt.xlabel('profit')
    plt.show()

profit_scatter('revenue')
```

**for the example i input the revenue column and as expected it has a very strong relation with the profit**
