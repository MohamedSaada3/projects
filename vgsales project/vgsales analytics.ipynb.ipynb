```python
companies=df["Publisher"].drop_duplicates()
company_names=list(companies)
print(company_names)
```

```python



# def sales_total():
d = []
for loop in company_names:
    df_test = df.loc[df['Publisher'] == loop]
    numbers = df_test.Global_Sales.sum()
    d.append(
        {'companies': loop, 'total_sales': numbers}
    )

df1 = pd.DataFrame(d)
    
    
nan_values=  df1[df1.isna().any(axis=1)]
df1.dropna(how='any',inplace=True)
df1.reset_index(drop=True,inplace=True)
    
df1    

```

```python

df1=df1.sort_values('total_sales',ascending=False)
df1
```
